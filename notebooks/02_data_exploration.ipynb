# Exploratory Data Analysis (EDA)

Load processed data

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("C:/Users/Navyashree/Documents/urban-air-quality-forecaster/data/processed/grid_air_quality.csv")
df.head()
```

```python
df.describe()
```

```python
# PM2.5 time trend

sample_grid = df.iloc[0]

grid_df = df[
    (df["grid_lat"]== sample_grid["grid_lat"]) &
    (df["grid_lon"]== sample_grid["grid_lon"])    
]

plt.figure()
plt.plot(grid_df["hour"],grid_df["pm25"])
plt.xlabel("Time")
plt.ylabel("PM2.5")
plt.title("PM2.5 Trend for a Grid Cell")
plt.xticks(rotation=0)
plt.show()
```

```python
# Pollution distribution

plt.figure()
plt.hist(df["pm25"],bins =20)
plt.xlabel("PM2.5")
plt.ylabel("Frequency")
plt.title("PM2.5 Distributiion")
plt.show()
```
