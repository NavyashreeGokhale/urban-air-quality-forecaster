Load tensors

```python
import numpy as np

X = np.load("../data/tensors/X.npy")
Y = np.load("../data/tensors/Y.npy")
adj_matrix = np.load("../data/processed/adjacency_matrix.npy")

print("X:", X.shape)
print("Y:", Y.shape)
print("Adj:", adj_matrix.shape)
```

Train–validation split (time-based)

```python
split = int(0.8 * X.shape[0])

X_train, X_val = X[:split], X[split:]  # past
Y_train, Y_val = Y[:split], Y[split:]  # future

print(X_train.shape, X_val.shape)
print(Y_train.shape, Y_val.shape)
```

Baseline model Linear Regression per node

```python
# Align X and Y in time
X = X[:-1]     # drop last timestep
Y = Y[1:]      # predict next timestep

#X[t] → predicts Y[t+1]
```

```python
# train–val split
split = int(0.8 * X.shape[0])

X_train = X[:split]
X_val   = X[split:]

Y_train = Y[:split]
Y_val   = Y[split:]
```

```python
# Node-level extraction

node = 0

Xn_train = X_train[:, node, :]
Xn_val   = X_val[:, node, :]

Yn_train = Y_train[:, node]
Yn_val   = Y_val[:, node]
```

#verify dataset sizes

```python
print("X shape:", X.shape)
print("Y shape:", Y.shape)

print("X_train:", X_train.shape)
print("X_val:", X_val.shape)

print("Y_train:", Y_train.shape)
print("Y_val:", Y_val.shape)
```

```python
Xn_val = X_val[:len(Y_val), node, :]
Yn_val = Y_val[:, node]
```

```python
model = LinearRegression()
model.fit(Xn_train, Yn_train)

preds = model.predict(Xn_val)

print("Yn_val:", Yn_val.shape)
print("preds:", preds.shape)

mse = mean_squared_error(Yn_val, preds)
print("Validation MSE:", mse)
```
